```python
# Dependencies and Setup
import pandas as pd

filepath_school = "Resources/schools_complete.csv"
filepath_student = "Resources/students_complete.csv"

school_df = pd.read_csv(filepath_school, encoding='UTF-8')
student_df = pd.read_csv(filepath_student, encoding='UTF-8')

raw_data = pd.merge(student_df, school_df, how="left", on=["school_name"])
raw_data.head()
```

```python
print(raw_data.shape)
raw_data.info()
```

```python
# Rename columns for dot formatting
raw_data.columns = [x.lower().replace(" ", "_") for x in raw_data.columns]
raw_data.head()
```

```python
##### District Summary
```

```python
# Total number of schools
school_count = raw_data.school_name.nunique()
school_count
```

```python
# Total number of students
student_count = raw_data.student_id.nunique()
student_count
```

```python
# Total budget
total_budget = raw_data.budget.unique().sum()
total_budget
```

```python
# Average math score
avg_math = raw_data.math_score.mean()
avg_math
```

```python
# Average reading score
avg_reading = raw_data.reading_score.mean()
avg_reading
```

```python
# Percentage of students who passed math (scores >= 70)
pass_math_count = raw_data[(raw_data.math_score >= 70)].count().student_id
pass_math_perc = pass_math_count / student_count
pass_math_perc
```

```python
# Percentage of students who passed reading (scores >= 70)
pass_reading_count = raw_data[(raw_data.reading_score >= 70)].count().student_id
pass_reading_perc = pass_reading_count / student_count
pass_reading_perc
```

```python
# Percentage of students that passed both math and reading
pass_math_reading_count = raw_data[(raw_data.math_score >= 70) & (raw_data.reading_score >= 70)].count().student_id
overall_pass_rate = pass_math_reading_count / student_count
overall_pass_rate
```

```python
# District summary DataFrame
district_summ = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "% Passing Math": [pass_math_perc],
    "% Passing Reading": [pass_reading_perc],
    "% Overall Passing": [overall_pass_rate]
})

district_summ["Total Students"] = district_summ["Total Students"].map("{:,}".format)
district_summ["Total Budget"] = district_summ["Total Budget"].map("${:,.2f}".format)
district_summ["% Passing Math"] = district_summ["% Passing Math"].map("{:.2%}".format)
district_summ["% Passing Reading"] = district_summ["% Passing Reading"].map("{:.2%}".format)
district_summ["% Overall Passing"] = district_summ["% Overall Passing"].map("{:.2%}".format)

district_summ
```
